The `prefiltered` was created by deleting all the JFR files that will be filtered out anyway. Therefore, it should be a
correct optimization.

```kotlin
val data = mapOf(
    "original" to load_data("jfrs/serialization/replication/prefiltered/interval_and_iteration_study/"),
    "new_base_with_memory" to load_data("jfrs/serialization/replication/prefiltered/new_base_with_memory/"),
    "new_base_with_memory-fixed" to load_data("jfrs/serialization/replication/prefiltered/new_base_with_memory-fixed/"),
    "control" to load_data("jfrs/serialization/replication/prefiltered/control/"),
    "control2" to load_data("jfrs/serialization/replication/prefiltered/control2/"),
)
```

```kotlin
fun extract_data(data: Data, f: String): List<Int> = data
    .filter { it.key["platform"]!! == "jvm" && it.key["sampleInterval"]!! == "100" && it.key["cycles"]!! == "100000" }
    .map { it.value[f]!! }
    .toList()
```

```kotlin
plot_histogram(data.mapValues { extract_data(it.value, "java.lang.String.substring") })
```

```kotlin
plot_histogram(data.mapValues { extract_data(it.value, "org.jetbrains.stdlibprofiling.TwitterBenchmark.encodeDecode") })
```

```kotlin
val filtered_data = data.filter { it.key != "new_base_with_memory-fixed" }
```

```kotlin
plot_histogram(filtered_data.mapValues { extract_data(it.value, "org.jetbrains.stdlibprofiling.TwitterBenchmark.encodeDecode") })
```

```kotlin
plot_histogram(filtered_data.mapValues { extract_data(it.value, "java.lang.String.substring") })
```

```kotlin
compareVarianceAndBias(filtered_data.mapValues { extract_data(it.value, "java.lang.String.substring").map { it.toDouble() } })
```

```kotlin
compareVarianceAndBias(filtered_data.mapValues { extract_data(it.value, "org.jetbrains.stdlibprofiling.TwitterBenchmark.encodeDecode").map { it.toDouble() } })
```

It seems that the big bias in the original measurement was cause by `-Xmx` and `-Xms` which we're incorrectly recorded
for replication and assumed to be ignored in the measurement. After discarding the measurement, the surviving bias seems
to be around the same order as the bias caused by AABB vs. ABAB measurement. We think that the root cause in difference
between the original measurement and control is the same as the difference between AABB and ABAB.
