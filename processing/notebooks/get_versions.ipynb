```kotlin
fun versionOf(clazz: Class<*>): String =
    clazz.`package`?.implementationVersion
        ?: clazz.protectionDomain?.codeSource?.location?.toString()
        ?: "<unknown>"

fun printVersion(label: String, className: String) {
    val v = runCatching { versionOf(Class.forName(className)) }
        .getOrElse { "<class not found: $className>" }
    println("$label: $v")
}
printVersion("DataFrame", "org.jetbrains.kotlinx.dataframe.DataFrame")
printVersion("Kandy", "org.jetbrains.kotlinx.kandy.ir.Plot")
printVersion("Statistics", "org.jetbrains.kotlinx.statistics.binning.BinsOption")
```
