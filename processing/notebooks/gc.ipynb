```kotlin
import java.io.File

fun loadData(path: String): Map<Map<String, String>, List<Double?>> = walkPath(path)
    .associateWith {
        it
            .substringAfterLast("/")
            .removeSuffix(".csv")
            .takeIf { it.contains("=") }
            ?.split(",")
            ?.map { it.split("=").let { (k, v) -> k to v } }
            ?.toMap()
    }
    .filter { it.value?.contains("name") ?: false }
    .map { it.value!! to
            File(it.key)
                .readText()
                .split(",")
                .map { it.toDoubleOrNull() }
    }
    .toMap()
```

```kotlin
// I suspect this is implemented incorrectly, as the results are more random than expected
// But maybe it's just how Otsu's method behaves
// Either way, it's faster for now to just find the split manually
fun otsu(data: List<Double>): Double {
    val sorted = data.sorted()
    return (1..sorted.size - 1)
        .associateWith { i ->
            val left = sorted.subList(0, i)
            val right = sorted.subList(i, sorted.size)
            // This formula was taken from wikipedia. It should be the product of weights which add to 1
            // When I divide all the values twice by sorted.size, I think I don't have enough Double precision
            // However, this division shouldn't be necessary as it as argmax (k * L) == argmax (L)
            (mean(left) - mean(right)).let { left.size * right.size * it * it }
        }
        .maxBy { it.value }!!
        .key
        .let { sorted[it] }
}
```

The main goal is to establish how much percent of the benchmark is spent in GC. This can be later used as a correction to the cleanly
measured benchmark assuming that the percent is the same in both methods.

Estimating this time proved harder than expected as the number of invocations per iteration is not constant in JMH.
However, time per iteration should be constant (this assumption is not 100% correct as JMH can't stop the benchmark mid-invocation).

The attempted method of estimation works as follows. Let $t$ denote the constant time per iteration. Let $a_i$ denote measured avgt results that didn't run GC and $b_i$ those which did.
We estimate the average time spent in GC / op as $E(b - E(a))$. Therefore, total time spent in GC can be estimated as $E(b - E(a))\sum{\frac{t}{b_i}}$. The total time is $nt$. In the ratio, the $t$ cancelles out, leaving the implemented formula.

For Native, the situation should be simple as the number of invocations per iteration $n$ is constant. This leads to the following formula, where $n$ cancels out

$$\frac{n\sum{\left(b_i - E(a)\right)}}{n\sum{a_i} + n\sum{b_i}}$$

```kotlin
if (jvmManualSplit.keys != nativeManualSplit.keys) throw Exception()

fun estimateGCPercWithConstT(data: List<Double>, threshold: Double): Double {
    val (noGC, GC) = data.partition { it < threshold }
    val avgTimeWoGC = mean(noGC)
    val avgGCTime = mean(GC.map { it - avgTimeWoGC })
    return avgGCTime * GC.sumOf { 1 / it} / data.size
}

fun estimateGCPercWithConstN(data: List<Double>, threshold: Double): Double {
    val (noGC, GC) = data.partition { it < threshold }
    val avgTimeWoGC = mean(noGC)
    return GC.sumOf { it - avgTimeWoGC } / data.sum()
}

val nativeGCPercBimodal = bimodalNative
    .map {
        val name = it.key["name"]!!
        val data = it.value
        val perc = nativeManualSplit[name]?.let { estimateGCPercWithConstN(data, it) } ?: 0.0
        name to perc
    }
    .toMap()

val jvmGCPercBimodal = loadJson("../../benchmarks/quantiles/jvm/results/all.json")
    .jsonArray.map { bm ->
        val name = bm.jsonObject["benchmark"]!!.jsonPrimitive.content.substringAfterLast(".")
        val perc = jvmManualSplit[name]?.let { estimateGCPercWithConstT(extractMetric(bm).first(), it) } ?: 0.0
        name to perc
    }
    .toMap()
```

This is very weird for Native as the GC mode is much higher the larger the substring. One would assume that therefore GC is more prevalent in larger substrings.
However, the processed data show the exact opposite.

While I'm very suspicios of the result,
this could be the case if GC would take much longer than the average iteration for smaller substring.
This seems to be the case as all the noGC iteration seems to be group tightly around "zero" and GC iteration are
quite far right on the x-axis. For larger substrings, all the data is much closer together.

Also, this could make sense at the number of objects is the same everywhere and GC shouldn't depend on the size of them too much.
However, the time to traverse/copy the string increases linearly with the size of the substring.

```kotlin
val nativeGCPercProf = GCprofNative.map { (params, data) ->
    val (GC, noGC) = data
        .zipWithNext()
        .partition { it.second == null }
        .let { (x, y) ->
            x.map { it.first!! } to y.mapNotNull { it.first }
        }
    params["name"]!! to (GC.sumOf { it - noGC.average() } / (GC.sum() + noGC.sum()))
}.toMap()
```

```kotlin
val jvmGCPercProf = loadJson("../../benchmarks/quantiles/jvm/results/allGCprof.json")
    .jsonArray.map { data ->
        val secondaryKeys = data.jsonObject["secondaryMetrics"]!!.jsonObject.keys.toList()
        val gcCountKey = secondaryKeys.single { it.contains("gc.count") }  // For some reason the field contains some weird special chars

        extractMetric(data).zip(extractMetric(data, gcCountKey)).map { (data, gcCount) ->
            data.zip(gcCount)
                .partition { it.second > 0.0 }
                .let { (x, y) -> x.map { it.first } to y.map { it.first} }
                .let { (GC, noGC) ->
                    GC.sumOf { it - noGC.average() } / (GC.sum() + noGC.sum())
                }
        }
            .let { data.jsonObject["benchmark"]!!.jsonPrimitive.content.substringAfterLast(".") to it }
    }
    .toMap()
    .also { println(it) }
    .mapValues { it.value.average() }
```

```kotlin
fun loadScores(json: JsonElement): Map<String, Double> = json
    .jsonArray.map {
        val name = it.jsonObject["benchmark"]!!.jsonPrimitive.content.substringAfterLast(".")
        val score = it.jsonObject["primaryMetric"]!!.jsonObject["score"]!!.jsonPrimitive.double
        name to score
    }
    .toMap()

val jvm = loadScores(loadJson("../../benchmarks/vanilla/report-stringsubstring-jvm.json"))
val native = loadScores(loadJson("../../benchmarks/vanilla/report-stringsubstring-linuxX64.json"))

if ( jvm.keys != native.keys) throw Exception()

val benchmarks = jvm.keys.toList()

fun Map<String, Double>.toNamedColumn(name: String): DataColumn<Double> = benchmarks
    .map { this[it]!! }
    .toColumn(name)

val df = dataFrameOf(
    benchmarks.toColumn("benchmark"),
    jvm.toNamedColumn("jvm"),
    native.toNamedColumn("native"),
    jvmGCPercBimodal.toNamedColumn("jvmGCPercBimodal"),
    nativeGCPercBimodal.toNamedColumn("nativeGCPercBimodal"),
    jvmGCPercProf.toNamedColumn("jvmGCPercProf"),
    nativeGCPercProf.toNamedColumn("nativeGCPercProf"),
)
```
