```kotlin
val mem = DataFrame.readCsv("mem.csv", header = listOf("virt", "res")).add("index") { it.index() }
```

```kotlin
mem.plot {
    line {
        x("index")
        y("virt")
        color = Color.BLUE
    }

    line {
        x("index")
        y("res")
        color = Color.RED
    }
}
```
