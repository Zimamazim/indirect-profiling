```kotlin
import jdk.jfr.consumer.*
import kotlin.io.path.Path
import java.time.Instant

var max_time = Instant.MIN

RecordingFile(Path("/home/Martin.Zimen/IdeaProjects/ktor/profiler_output/jvm-URLBuilderTest.testPathSegments.jfr")).use {
    while (it.hasMoreEvents()) {
        it.readEvent()!!
            .also { if (it.startTime > max_time) max_time = it.startTime }
            .takeIf { it.eventType.name == "jdk.ExecutionSample" }
            ?.takeIf { it.stackTrace?.frames?.any { frame -> frame.method.name.contains("collectionSizeOrDefault") } ?: false }
            ?.let {
                println(it.startTime)
            }
    }
}

println("MAX: $max_time")
```

This was used to test hypothesis that `collectionSizeOrDefault` is optimized out by C1/C2. The samples seem to be even
throughout the whole execution. Therefore, it doens't seem to be the case.
