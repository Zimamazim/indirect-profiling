```kotlin
import jfr_processor.profilelib.*

val project_root = "/home/Martin.Zimen/IdeaProjects/ktor/"
val profile_root = project_root + "profiler_output_weekend/"
```

```kotlin
val native_debug_map = ktor_get_native_debug_map(project_root)
val jvm_fun_to_src_map = ktor_get_fun_to_src_map(project_root)
```

```kotlin
val jvm_freq = mutableMapOf<String, Int>()
val native_freq = mutableMapOf<String, Int>()

get_valid_tests(profile_root).forEach { testname ->
    lazy_samples(profile_root + "jvm-" + testname + ".jfr")
        .mapNotNull { truncate_stack_trace(it) { jvm_get_name(it) == testname } }
        .filter { jvm_is_sample_platform_independent(it, jvm_fun_to_src_map) }
        .toFreq(jvm_freq, ::jvm_get_name)

    lazy_samples(profile_root + "linuxX64-" + testname + ".jfr")
        .mapNotNull { truncate_stack_trace(it) { native_get_name(it)?.contains(testname) ?: false } }
        .filter { native_is_sample_platform_independent(it, native_debug_map) }
        .toFreq(native_freq, ::native_get_name)
}
```

```kotlin
match_and_compare_freqs(jvm_freq, native_freq) { it.second.jvm_count > 2 && it.second.native_count > 2 }
```
