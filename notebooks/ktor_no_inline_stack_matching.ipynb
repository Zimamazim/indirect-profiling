```kotlin
import jfr_processor.profilelib.*
val root = "/home/Martin.Zimen/temp_root/"
```

```kotlin
val jvm_fun_names = mutableSetOf<String>()
val native_fun_names = mutableSetOf<String>()

lazy_function_names(root, "jvm").filter { it.startsWith("kotlin.") }.forEach { jvm_fun_names.add(it) }
lazy_function_names(root, "linuxX64").filter { it.startsWith("kotlin.") }.forEach { native_fun_names.add(it) }
```

```kotlin
val intersection = jvm_fun_names intersect native_fun_names
println(jvm_fun_names)
println()
println(native_fun_names)
println()
println(intersection)
```

```kotlin
val jvm_samples = get_valid_tests(root)
    .map { test ->
        lazy_samples(
            root + "jvm-" + test + ".jfr",
            ::jvm_process_frame
        ).map { remove_base(it, test) }.filterNotNull()
    }
    .flatten()
    .filter { it.any { get_name(it) == "kotlin.collections.CharIterator.next" } }
    .map { it.map { get_name(it) } }
    .toSet()
```

```kotlin
val native_samples = get_valid_tests(root)
    .map { test ->
        lazy_samples(
            root + "linuxX64-" + test + ".jfr",
            ::native_process_frame
        ).map { remove_base(it, test) }.filterNotNull()
    }
    .flatten()
    .filter { it.any { get_name(it) == "kotlin.collections.CharIterator.next" } }
    .map { it.map { get_name(it) } }
    .toSet()
```

```kotlin
println("JVM")
jvm_samples.forEach { println(it) }
println("Native")
native_samples.forEach { println(it) }
println("Intersection")
(jvm_samples intersect native_samples).forEach { println(it) }
```
